```python
import pandas as pd

# Replace with your Parquet file path
file_path = "Dataset/Lyft_2021.parquet"
sampling_fraction = 0.1

# Define the data types for each column
dtype_dict = {
    'pickup_datetime': 'string',
    'dropoff_datetime': 'string',
    'pulocationid': 'int64',  # Assuming pulocationid is an integer
    'dolocationid': 'int64',  # Assuming dolocationid is an integer
    'trip_miles': 'float64',
    'driver_pay': 'float64',
    'puzone': 'string',
    'puborough': 'string',
    'dozone': 'string',
    'doborough': 'string',
}

# Read the Parquet file into a DataFrame
df = pd.read_parquet(file_path, engine='pyarrow')

# Sample the DataFrame
sampled_df = df.sample(frac=sampling_fraction, random_state=42)

# Save the sampled DataFrame to a new Parquet file
sampled_file_path = "Dataset/Lyft_2021.parquet"
sampled_df.to_parquet(sampled_file_path, engine='pyarrow', compression='snappy')

print(f"Original dataset size: {len(df)} rows")
print(f"Sampled dataset size: {len(sampled_df)} rows")
print(f"Sampled dataset saved to: {sampled_file_path}")
```

```python
import pandas as pd
from collections import Counter

df_2021 = pd.read_parquet(r'backend/Dataset/2021.parquet')
df_2022 = pd.read_parquet(r'backend/Dataset/2022.parquet')

df_2021 = df_2021.drop(['PUZone', 'PUBorough', 'DOZone', 'DOBorough'], axis=1)
df_2022 = df_2022.drop(['PUZone', 'PUBorough', 'DOZone', 'DOBorough'], axis=1)

def process_year(df):
    daily_data = {}

    for date in pd.to_datetime(df['pickup_datetime']).dt.date.unique():
        day_df = df[pd.to_datetime(df['pickup_datetime']).dt.date == date]

        most_common_pulocation = Counter(day_df['PULocationID']).most_common(1)[0][0]
        most_common_dolocation = Counter(day_df['DOLocationID']).most_common(1)[0][0]

        total_trip_miles = day_df['trip_miles'].sum()
        total_driver_pay = day_df['driver_pay'].sum()

        daily_data[date] = {
            'PULocationID': most_common_pulocation,
            'DOLocationID': most_common_dolocation,
            'total_trip_miles': total_trip_miles,
            'total_driver_pay': total_driver_pay
        }

    return pd.DataFrame.from_dict(daily_data, orient='index')

df_2021_daily = process_year(df_2021)
df_2022_daily = process_year(df_2022)

df_all_daily = pd.concat([df_2021_daily, df_2022_daily])

df_2023 = df_all_daily.groupby(df_all_daily.index).mean().reset_index()
df_2023 = df_2023.rename(columns={'index': 'date'})
df_2023['date'] = pd.to_datetime(df_2023['date'])
print(df_2023['date'].dtype)
df_2023_meow = df_2023
df_2023_meow['date'] = df_2023_meow['date'].apply(lambda x: x.replace(year = x.year + 2))
print(df_2023_meow)
```

```python
df_2023_meow.to_parquet("backend/Dataset/2023.parquet")
df_2023_meow.to_csv("backend/Dataset/2023.csv")
print(df_2023_meow)
```

```python
import pandas as pd
df = pd.read_csv("backend/Dataset/2023.csv")
import matplotlib.pyplot as plt

# Convert the 'date' column to datetime format
df['date'] = pd.to_datetime(df['date'])

# Plot total_trip_miles over time
plt.figure(figsize=(12, 6))
plt.plot(df['date'], df['total_trip_miles'])
plt.xlabel('Date')
plt.ylabel('Total Trip Miles')
plt.title('2023 Forecased Trip Miles')
plt.xticks(rotation=45)
plt.grid()
plt.show()

# Plot total_driver_pay over time
plt.figure(figsize=(12, 6))
plt.plot(df['date'], df['total_driver_pay'])
plt.xlabel('Date')
plt.ylabel('Total Driver Pay')
plt.title('2023 Forecasted Revenue')
plt.xticks(rotation=45)
plt.grid()
plt.show()
```
